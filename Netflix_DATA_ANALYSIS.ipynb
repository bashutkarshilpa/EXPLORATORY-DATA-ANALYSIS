```python
cntry_plot=sns.countplot(y='country',order=netflix_df['country'].value_counts().head(10).index[0:11],data=netflix_df)
total=len(netflix_df)
for patch in cntry_plot.patches:
    count=patch.get_width()
    y=patch.get_y()+0.5
    x=patch.get_x()
    cntry_plot.annotate(count,(x,y))
    
plt.title('Contrywise analysis')   
```
