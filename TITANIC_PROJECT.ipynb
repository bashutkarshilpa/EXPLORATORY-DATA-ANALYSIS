## Annotating:

```python
plot=sns.countplot(x='Pclass',data=titanic_df)
for patch in plot.patches:
    count=patch.get_height()
    x=patch.get_x()+0.4
    y=count/2
    plot.annotate(count,(x,y))
```

## % of passengers in each class along with no.of passengers:

```python
plot=sns.countplot(x='Pclass',data=titanic_df)
total=len(titanic_df)
for patch in plot.patches:
    count=patch.get_height()
    x=patch.get_x()+0.1
    y=count/2
    per=count/total*100
    percentage="{:.2f}%".format(per)
    annotation=f"{count}({percentage})"
    plot.annotate(annotation,(x,y))
```

### Conclusion:
- No.of passengers in Pclass-1: 216 and % of passengers in Pclass-1 is :24.24%
- No.of passengers in Pclass-2: 184 and % of passengers in Pclass-1 is :20.65%
- No.of passengers in Pclass-2: 491 and % of passengers in Pclass-1 is :55.11%

- Therefore,from this we can say that the passengers travelled more in Pclass-3 i.ee., >50% 
- Less people travelled in Pclass-2 (20.65%) only

### Conclusion:
- No.of passengers survived in Pclass-1: 136
- No.of passengers survived in Pclass-2: 87

```python
plot1=sns.countplot(x='Pclass',hue='Survived',data=titanic_df)
i=1
j=1
for patch in plot1.patches[0:3]:
    total=len(titanic_df[titanic_df['Pclass']==i])
    count=patch.get_height()
    x=patch.get_x()-0.1
    y=count/2
    per=count/total*100
    percentage='{:.2f}%'.format(per)
    value=f"{count}({percentage})"
    plot1.annotate(value,(x,y))
    i=i+1
for patch in plot1.patches[3:]:
        total=len(titanic_df[titanic_df['Pclass']==j])
        count=patch.get_height()
        x=patch.get_x()
        y=count/2
        per=count/total*100
        percentage='{:.2f}%'.format(per)
        value=f"{count}({percentage})"
        plot1.annotate(value,(x,y))
        j=j+1
        
```

## Conclusion:
-  survival rate of passengers in Pclass-1: 62.96%
-  survival rate of passengers in Pclass-2: 47.28%
-  survival rate of passengers in Pclass-3: 24.24%
    
- Therefore, passengers in Pclass-1  have highest survival rate (62.96%)
- passenegers in Pclass-3 have least survival rate (24.24%)

## Handling Missing values or Pre-Processing:

```python
titanic_df.info()
```

### No.of null values in each column:

```python
titanic_df.isna().sum()
```

- Age : 177 missing values
- cabin :687 missing values
- embarked:2 missing values

## filling missing values :

```python
df1=titanic_df.fillna({'Age':titanic_df['Age'].mean(),'Embarked':titanic_df['Embarked'].mode().loc[0]})
```

```python
df1
```

### droping cabin column
- as >50% data of cabin column is missing

```python
df1.drop('Cabin',axis=1,inplace=True)
```

```python
df1
```

```python
df1.info()
```

```python
df1.isna().sum()
```

- ML can't take object values so pre-processing it with numerical values

```python
titanic_df.describe(include=object)
```

- Name,sex,Ticket,Cabin,Embarked are object columns
- drop Name,Ticket,Cabin 
- refill sex and embarked with numerical stuff

```python
df2=df1.drop(['Name','Ticket'],axis=1)
```

```python
df2
```

```python
encoded_sex=pd.get_dummies(df2['Sex'])
```

```python
encoded_sex
```

```python
df3=pd.concat([df2,encoded_sex],axis=1)
```

```python
df3
```

```python
encoded_embarked=pd.get_dummies(df3['Embarked'])
```

```python
encoded_embarked
```

```python
df5=pd.concat([df3,encoded_embarked],axis=1)
```

```python
df5
```

```python
df5.drop(['Sex','Embarked'],axis=1,inplace=True)
```

```python
df5
```

```python
pre_processed_titanic_df=df5
```

```python
pre_processed_titanic_df
```

### Conclusion:
- data got pre_processed and ready to train ML
